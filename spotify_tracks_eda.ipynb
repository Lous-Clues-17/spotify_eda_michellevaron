```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#loading in dataset and initial inspection
df = pd.read_csv('dataset.csv')

print(df.shape)
df.head(10)
```

```python
#dropping unnamed:0 column
df.columns

df = df.drop('Unnamed: 0', axis = 1)

df.head(10)
```

```python
#checking column data types
df.dtypes
```

```python
#finding missing values
df.isna()

df.isna().sum()
```

```python
#locating row with missing column values
df.loc[df['artists'].isna()]
```

```python
#finding duplicated data
df.duplicated(keep=False).sum()
```

```python
#locating duplicated data rows
df.loc[df.duplicated(keep=False)].sort_values('track_id')
```

```python
#locating rows where the track id column values are duplicated
df.loc[df['track_id'].duplicated(keep=False)].sort_values('track_id')
```
