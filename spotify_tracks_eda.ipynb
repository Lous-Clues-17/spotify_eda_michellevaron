```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#loading in dataset and initial inspection
df = pd.read_csv('dataset.csv')

print(df.shape)
df.head(10)
```

```python
#dropping unnamed:0 column
df.columns

df = df.drop('Unnamed: 0', axis = 1)

df.head(10)
```

```python
#checking column data types
df.dtypes
```

```python
#finding missing values
df.isna()

df.isna().sum()
```

```python
#locating row with missing column values
df.loc[df['artists'].isna()]
```

```python
#finding duplicated data
df.duplicated(keep=False).sum()
```

```python
#locating duplicated data rows
df.loc[df.duplicated(keep=False)].sort_values('track_id')
```

```python
#locating rows where the track id column values are duplicated
df.loc[df['track_id'].duplicated(keep=False)].sort_values('track_id')
```

```python
#univariate analysis of qualitative variables 
df['artists'].value_counts()
```

```python
#univariate analysis of quanitative variables 
df.describe()
```

```python
#plotting danceability distribution as histogram
df['danceability'].hist(bins=60)
```

```python
#plotting danceability distribution as box plot
sns.boxplot(x=df['danceability'])
```

```python
#multivariate analysis begins by reducing dataframe 
#to quantitaive variables only and displaying correlation matrix
df.columns
df_corr =df[['popularity',
       'duration_ms', 'danceability', 'energy', 'key', 'loudness',
       'mode', 'speechiness', 'acousticness', 'instrumentalness', 'liveness',
       'valence', 'tempo', 'time_signature']].corr()
df_corr
```

```python
#creating heat map 
fig, ax = plt.subplots(figsize=(15,15))
sns.heatmap(df_corr,annot=True, cmap='mako')
```

```python
#creating pair plot
sns.pairplot(df[['acousticness', 'energy' ]], 
             height = 5, aspect = 1.5 )
```

```python
#creating another pair plot with smaller random sample of 5,000 data points 
df_sample = df[['acousticness', 'energy']].sample(n=5000, random_state=1)

sns.pairplot(df_sample[['acousticness', 'energy' ]], 
             height = 5, aspect = 1.5, diag_kind='kde' )
```
